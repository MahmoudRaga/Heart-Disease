```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
plt.style.use ("dark_background")
```

```python
dataset = pd.read_csv ('heart.csv')
dataset
```

```python
dataset.info()
dataset.nunique()
```

```python
dataset.drop(['oldpeak'], axis=1, inplace=True)
dataset
```

```python
x = dataset.iloc[:,:-1].values
x
```

```python
y = dataset.iloc[:,-1].values
y
```

```python
SC = StandardScaler()
x[:,0:1] = SC.fit_transform(x[:,0:1])
x[:,3:5] = SC.fit_transform(x[:,3:5])
x[:,7:8] = SC.fit_transform(x[:,7:8])
```

```python
data_x = pd.DataFrame (x ,columns = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','slope','ca','thal'])
data_x
```

```python
data_y = pd.DataFrame (y ,columns = ['target'])
data_y
```

```python
x_train,x_test,y_train,y_test = train_test_split (x ,y ,test_size = 0.2 ,random_state = 0)
```

```python
svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
df_comp = pd.DataFrame({'Actual':y_test , 'Predict':y_pred})
df_comp
```

```python
plt.title (' Actual & Predict ',color = 'w')
sns.heatmap(df_comp)
sns.pairplot(df_comp)
```

```python
df_test = pd.DataFrame(y_test)
df_pred = pd.DataFrame(y_pred)

print ('y_test = ',df_test.sum())
print ('y_pred = ',df_pred.sum())
```
